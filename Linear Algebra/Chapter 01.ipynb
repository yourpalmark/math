Every vector is associated with a single pair of numbers `v = (x, y)`  
Vectors are commonly displayed as $v = \begin{bmatrix} x \\ y \end{bmatrix}$

````python
v = (-2, 3)

x_line = [(0, 0), (-2, 0)]
y_line = [(-2, 0), v]

draw2D(
    Arrow2D(v, color='yellow', label="v"),
    Line2D(*x_line, color='green', label="x"),
    Line2D(*y_line, color='orange', label="y")
)
````

**INFO**  
That's true with 3D vectors as well: v = (x, y, z)

````python
v = (-2, 3, 3)

x_line = [(0, 0, 0), (-2, 0, 0)]
y_line = [(-2, 0, 0), (-2, 3, 0)]
z_line = [(-2, 3, 0), v]

draw3D(
    Arrow3D(v, color='yellow'),
    Line3D(*x_line, color='green'),
    Line3D(*y_line, color='orange'),
    Line3D(*z_line, color='blue'),
    azim=-140, elev=25
)
````

### Linear Algebra

````python
v = (1, 2)
w = (3, -1)

draw2D(
    Arrow2D(v, color='yellow'),
    Arrow2D(w, color='pink'),
    grid=True
)
````

#### Addition
Walk along the length of vector v and then the length of vector w and you end up at vector vw  
```
v = (1, 2)
w = (3, -1)
vw = (4, 1)
```
That's the same as 1 + 3 and 2 + (-1)

````python
v = (1, 2)
w = (3, -1)
vw = add(v, w)

draw2D(
    Arrow2D(v, color='yellow'),
    Arrow2D(vw, v, color='pink'),
    Arrow2D(vw, color='orange', linestyle=(5,(3,6))),
    grid=True
)
````

#### Scaling (Multiplication by a Number)

````python
v = (3, 1)

draw2D(
    Arrow2D(v, color='yellow'),
    grid=True
)
````

```
v = (3, 1)
v_s1 = scale(2, v)
```
That's the same as multiplying each component of the vector by 2: `(3 x 2)` and `(1 x 2)`

````python
v = (3, 1)
v_s1 = scale(2, v)

draw2D(
    Arrow2D(v, color='yellow'),
    Arrow2D(v_s1, color='yellow'),
    grid=True
)
````

````python
v = (3, 1)
v_s2 = scale(1/3, v)

draw2D(
    Arrow2D(v, color='yellow'),
    Arrow2D(v_s2, color='yellow'),
    grid=True
)
````

````python
v = (3, 1)
v_s3 = scale(-1.8, v)

draw2D(
    Arrow2D(v, color='yellow'),
    Arrow2D(v_s3, color='yellow'),
    grid=True
)
````
