The unit vector in the x direction with length 1 is known as: $\mathbf{\hat{i}}$  
The unit vector in the y direction with length 1 is known as: $\mathbf{\hat{j}}$  
A vector of $\begin{bmatrix} 3 \\ -2 \end{bmatrix}$ is stretching $\hat{i}$  by 3 and $\hat{j}$  by -2  

```python
i = (1, 0)
j = (0, 1)
s_i = (3, 0)
s_j = (0, -2)
v = (3, -2)

draw2D(
    Arrow2D(i, color='green', label=r"$\mathbf{\hat{i}}$"),
    Arrow2D(j, color='orange', label=r"$\mathbf{\hat{j}}$"),
    Arrow2D(s_i, color='green'),
    Arrow2D(s_j, color='orange'),
    Arrow2D(v, color='yellow'),
    grid=True
)
```
