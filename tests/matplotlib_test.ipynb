### Points

```python
pts = [(6,4,'red','red'), (3,1), (1,2)]

draw2D(
    Point2D(2, 3, color="white", label="x"),
    Points2D(*pts, color='white')
)
```

```python
pts = [(6,4,4,'red','red'), (3,1,0), (1,2,-2)]

draw3D(
    Point3D(2, 3, 2, color="white", label="x"),
    Points3D(*pts, color='white')
)
```

### Lines

```python
v = (-2, 3)
w = (-1, 1)

draw2D(
    Line2D(v, w, color='green', label='vw')
)
```

```python
v = (-2, 3, 3)
w = (-1, 1, 2)

draw3D(
    Line3D(v, w, color='green', label='vw'),
    azim=-140, elev=25
)
```

### Arrows

```python
v = (-2, 3)

draw2D(
    Arrow2D(v, color='yellow', label='v')
)
```

```python
v = (-2, 3, 3)

draw3D(
    Arrow3D(v, color='yellow', label='v'),
    azim=-140, elev=25
)
```
