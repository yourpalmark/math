```python
from manim import *
```

```python
%%manim -v WARNING --disable_caching -ql -s VectorCoordinateLabel

class VectorCoordinateLabel(Scene):
    def construct(self):
        config.pixel_width = 300
        
        plane = NumberPlane()

        v1 = [1, 2]
        v2 = [-3, -2]

        vec1 = Vector(v1, color=GREEN)
        vec2 = Vector(v2, color=BLUE)
        lab1 = vec1.coordinate_label(color=GREEN)
        lab2 = vec2.coordinate_label(color=BLUE)

        self.add(plane, vec1, vec2, lab1, lab2)
```

```python
%%manim -v WARNING --disable_caching -ql -s VectorAddition

class VectorAddition(Scene):
    def construct(self):
        config.pixel_width = 300

        plane = NumberPlane()
        
        v1 = [1, 2]
        v2 = [-3, -2]

        # Create vector arrows
        vec1 = Vector(v1, color=GREEN)
        vec2 = Vector(v2, color=BLUE)

        # Create the sum vector
        v3 = [v1[0] + v2[0], v1[1] + v2[1]]
        vec3 = Vector(v3, color=RED)

        # Add labels
        lab1 = MathTex("\\vec{v}_1", color=GREEN).next_to(vec1.get_center(), RIGHT)
        lab2 = MathTex("\\vec{v}_2", color=BLUE).next_to(vec2.get_center(), UP)
        lab3 = MathTex("\\vec{v}_1 + \\vec{v}_2", color=RED).next_to(vec3.get_center(), DOWN)

        def label_updater(mob):
            mob.next_to(vec2.get_center(), UP)
        lab2.add_updater(label_updater)
        
        self.add(plane, vec1, vec2, vec3, lab1, lab2, lab3)
        
        self.play(
            vec2.animate.put_start_and_end_on(vec1.get_end(),vec3.get_end())
        )
```
