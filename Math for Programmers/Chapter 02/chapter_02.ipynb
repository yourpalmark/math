# Chapter 2: Drawing with 2D Vectors

```python
from manim import *
```

```python
%%manim -v WARNING --disable_caching -ql -r 600,600 -s ex_2_2

from number_grid import *

class ex_2_2(Scene):
    def construct(self):
        number_grid = NumberGrid(-2, 4, 1, -4, 2, 1, include_tips=False)

        o_vec = [0, 0]
        vec = [2, -2]

        origin = Dot(
            point=number_grid.grid.c2p(*o_vec),
            radius=0.1,
        )
        point = Dot(
            point=number_grid.grid.c2p(*vec),
            radius=0.1,
        )

        vector = Arrow(
            start=number_grid.grid.c2p(*o_vec),
            end=number_grid.grid.c2p(*vec),
            buff=0.1,
            color=RED,
        )

        self.add(number_grid.group, origin, point, vector)
```

```python
%%manim -v WARNING --disable_caching -ql -s ex_2_5

from number_grid import *

class ex_2_5(Scene):
    def construct(self):
        number_grid = NumberGrid(-11, 11, 1, -10, 110, 10, include_tips=False)

        dots = [Dot(number_grid.grid.c2p(x,x**2,0), color=WHITE) for x in range(-10,11)]

        dot_group = VGroup(*dots)

        self.add(number_grid.group, dot_group)
```
